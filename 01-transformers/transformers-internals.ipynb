```python
# understaing topk from torch sampling
import torch
```

```python
sentence = "I learn Machine Learning every"
sentence_tokens = tokenizer(sentence, return_tensors="pt")["input_ids"]
model_output = model(sentence_tokens).logits[0, -1]
op_logits = model_output
top_k_values, top_k_indices = torch.topk(model_output, k=5)
print("Top-k token IDs:", top_k_indices) #their token IDs in vocab
print("Top-k token probabilities:", top_k_values.softmax(dim=0)) #the 5 highest raw scores to probability
```

```python
for index in top_k_indices:
    decoded_token = tokenizer.decode([index])
    print(decoded_token)
```

```python
#sampling methods
# 1. Greedy Sampling: Selects the token with the highest probability at each step, which can lead to repetitive and less diverse outputs.
# 2. Top-k Sampling: Selects the next token from the top k most probable tokens, introducing some randomness while still focusing on likely options.
# 3. Top-p (Nucleus) Sampling: Selects the next token from the smallest set of tokens whose cumulative probability exceeds a threshold p, 
# allowing for more dynamic and contextually relevant outputs.
```

```python
def greedy_decode(logits):
    return logits.argmax()
    
```

```python
next_token = greedy_decode(op_logits)
human_readable_next_token = tokenizer.decode([next_token])
print("Greedy decoded next token:", human_readable_next_token)
```

```python
def top_k_sampling(logits, k):
    """Samples the next token from the top k most probable tokens based on the provided logits."""
    top_k_values, top_k_indices = torch.topk(logits, k) # values = raw scores, indices = token ids
    top_k_probabilities = top_k_values.softmax(dim=0) # normalizing using softmax
    sampled_index = torch.multinomial(top_k_probabilities, num_samples=1)
    return top_k_indices[sampled_index]
```

```python
next_op_token = top_k_sampling(op_logits, k=50)
human_readable_next_op_token = tokenizer.decode([next_op_token])
print("Top-k sampled next token:", human_readable_next_op_token)
```

```python
def top_p_sampling(logits, p): # nucleus sampling
    """
    sort tokens by probabilities, keep smallest number whose cumulative probability exceeds p, and sample from them.
    """
    sorted_logits, sorted_indices = torch.sort(logits, descending=True)
    sorted_prob = sorted_logits.softmax(dim=0)
    cumulative_prob = torch.cumsum(sorted_prob, dim=0)

    # mask tokens outside nucleus
    mask = cumulative_prob > p
    sorted_prob[mask] = float('-inf') # set to -inf so that they have 0 probability after softmax

    # sample from filtered logits
    filtered_prob = sorted_prob.softmax(dim=0)
    sampled_index = torch.multinomial(filtered_prob, num_samples=1)

    # return the original token index of the sampled token
    return sorted_indices[sampled_index]
```

```python
top_p_token = top_p_sampling(op_logits, p=0.9)
human_readable_top_p_token = tokenizer.decode([top_p_token])
print("Top-p sampled next token:", human_readable_top_p_token)
```

```python
# Temperature Sampling: Adjusts the probabilities of the next tokens by a temperature parameter, allowing for more or less randomness in the output. 
# A higher temperature results in more random outputs, while a lower temperature makes the model more confident and deterministic in its choices.

def temperature_sampling(logits, temperature):
    sclaed_logits = logits / temperature
    probabilities = sclaed_logits.softmax(dim=0)
    sampled_index = torch.multinomial(probabilities, num_samples=1)
    return sampled_index
```

```python
temperature_token = temperature_sampling(op_logits, temperature=0.7)
human_readable_temperature_token = tokenizer.decode([temperature_token])
print("Temperature sampled next token:", human_readable_temperature_token)
```

```python
def random_sampling(logits):
    """Randomly samples the next token based on the probabilities derived from the logits."""
    probabilities = logits.softmax(dim=0)
    sampled_index = torch.multinomial(probabilities, num_samples=1)
    return sampled_index
```

```python
random_token = random_sampling(op_logits)
human_readable_random_token = tokenizer.decode([random_token])
print("Randomly sampled next token:", human_readable_random_token)
```

```python
## comparing all sampling methods

sentence = "Today is the day "
tokenized_sentence = tokenizer(sentence, return_tensors="pt")
modal_op = model(**tokenized_sentence)
logits = modal_op.logits[0, -1]


# Greedy
next_token = greedy_decode(op_logits)
print("Greedy:", tokenizer.decode([next_token.item()]))

# Top-k
next_token = top_k_sampling(op_logits, k=50)
print("Top-k:", tokenizer.decode([next_token.item()]))

# Top-p
next_token = top_p_sampling(op_logits, p=0.9)
print("Top-p:", tokenizer.decode([next_token.item()]))

# Temperature
next_token = temperature_sampling(op_logits, temperature=0.7)
print("Temperature:", tokenizer.decode([next_token.item()]))

# Random
next_token = random_sampling(op_logits)
print("Random:", tokenizer.decode([next_token.item()]))
```

```python
# Different sampling strategies are employed to manipulate the probability distrubution of the token-pools, allowing for more control over the generated text's creativity and coherence.

sentence = "the sun is the vernal equin"
token_ids = tokenizer(sentence, return_tensors="pt")["input_ids"]
model_output = model(token_ids).logits[0, -1]
probabilities = model_output.softmax(dim=0)
top10 = torch.topk(probabilities, k=10)

for index, value in zip(top10.indices, top10.values):
    decoded_token = tokenizer.decode([index])
    print(f"Token ID: {index.item()} -> Decoded Token: '{decoded_token}' with probability {value.item():.1%}'")
```
